```python
# Присваиваем признак вида выборки и склеиваем в один датасет для трансформации
train_data.loc[:, 'sample'] = 'train'
test_data.loc[:, 'sample'] = 'test'
df = train_data.append(test_data).reset_index(drop=True)
```

```python
# смотри каково разнообразие значений в категориальных переменных
for i in df.columns: # перебираем все столбцы
    if (str(train_data[i].dtype) == 'object') &(i not in ['_id', 'sample']): # если тип столбца - object и это не идентификатор
        print('='*10)
        print(i) # выводим название столбца
        print(set(train_data[i])) # выводим все его уникальные значения 
        print('\n') # выводим пустую строку
```

```python
def preproc_data(df_input):
    ''' Функция препроцессинга '''
    df_output = df_input.copy()
    
    # определяем категориальные признаки
    cat_features=[col for col in df_output.columns if str(df_output[col].dtype) == 'object']
    cat_features.remove('sample') # исключаем признак выборки
    cat_features.remove('_id') # исключаем идентификатор

    # "приклеиваем" дамми-признаки столюцами справа
    df_output=pd.concat([df_output, pd.get_dummies(df_output[cat_features])], axis=1)
    # удаляем исходный категориальные признаки
    df_output.drop(cat_features, axis=1, inplace=True)
    
    return df_output
```

```python
# Применяем функцию препроцессинга
df_preproc = df.pipe(preproc_data)

# Разделяем выборки
df_train_preproc = df_preproc.query('sample == "train"').drop(['sample'], axis=1)
df_test_preproc = df_preproc.query('sample == "test"').drop(['sample'], axis=1)
```

```python
# Выделяем target
y = df_train_preproc['target']
# Делим признаки на train и тест
X, X_test = df_train_preproc.drop(['target','_id'], axis=1), df_test_preproc.drop(['target','_id'], axis=1)
# выделяем вектор идентификаторов тестовой выборки
test_id=df_test_preproc['_id']
```

```python
from sklearn.metrics import roc_auc_score, roc_curve
from sklearn.model_selection import StratifiedKFold
from sklearn.tree import DecisionTreeClassifier

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import RobustScaler

from sklearn.model_selection import GridSearchCV
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import StratifiedKFold

from scipy.stats import randint as randint
from scipy.stats import uniform
```

## Модель №1 - Деревья решений

```python
# Определим пространство поиска
param_grid = {
    'criterion': ['gini', 'entropy'],
    'max_depth': randint(2, 10),
    'min_samples_leaf': randint(1, 10),
    'class_weight': [None, 'balanced']}

# Разбивка на фолды
cv = StratifiedKFold(n_splits=5, shuffle=True, random_state=123)

# Будем делать 100 запусков поиска
model = DecisionTreeClassifier(random_state=123)
random_search = RandomizedSearchCV(model, param_distributions=param_grid, n_iter=100, n_jobs=-1,
                                   cv=cv, scoring='roc_auc', random_state=123)
# Ищем
random_search.fit(X, y)
```

```python
# лучшее качество
random_search.best_score_
```

```python
# лучшие параметры
random_search.best_params_
```

## Уже только это решение дало в baseline 0.93128 на public leaderboard

```python
# Далее решил обучить модель с данным и параметрами не на кросс-валидации, а на всей выборке (не знаю зачем поменял random-state))
model = DecisionTreeClassifier(criterion='entropy', max_depth=6, min_samples_leaf=10, class_weight='balanced',random_state=42)
model.fit(X, y)
```

```python
# Получаем предсказания
fitted=model.predict_proba(X)[:,1]
predictions=model.predict_proba(X_test)[:,1]
```

```python
# Посмотрим ROC-AUC на обучении на всякий случай (может переобучились вдруг...)
roc_auc_score(y, fitted)
```

```python
# Запусивыаем результаты
result=pd.DataFrame(columns=['_id', 'target'])
result['_id']=test_id
result['target']=predictions
result.to_csv('submission.csv', sep=',', index=False)
```
